```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/reviews.csv")
```

```python
df.head()
```

```python
def transform_data(df: pd.DataFrame):
    prof_name = df["Professor_Name"]
    review = df["Comment"]
    
    return f"{prof_name}: {review}"
```

```python
df = df.apply(transform_data, axis=1)
df = pd.DataFrame(df)
```

```python
df.to_csv("data/data.csv", index=False)
```
